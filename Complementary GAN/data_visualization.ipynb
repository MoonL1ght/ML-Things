````python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
def show_classes_ratio(data):
    data_classes = data.groupby('CLASS').size()
    neg_classes = data_classes[0]*100/(data_classes[0]+data_classes[1])
    pos_classes = data_classes[1]*100/(data_classes[0]+data_classes[1])
    print('Data: NEG classes %d (%.4f%%), POS classes %d (%.4f%%)' %(data_classes[0], neg_classes,
                                                                     data_classes[1], pos_classes))
    labels = ['NEG', 'POS']
    sizes = [neg_classes, pos_classes]

    figure = plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%.1f%%', shadow=False, startangle=90)
    plt.axis('equal')

    plt.show()
````

<center><h3>Visualizing 2D shperical data</h3></center>
Generating script:
```
python generate_data.py 10000 2 0.3
```

````python
DATA_FILE = './data/data_10000_2dim.csv'
data_2d = pd.read_csv(DATA_FILE, sep=';').drop(['Unnamed: 0'], axis=1)

show_classes_ratio(data_2d)
````

````python
data_2d = np.array(data_2d)
pos_data = data_2d[data_2d[:, 2] == 1][:100]
neg_data = data_2d[data_2d[:, 2] == 0][:100]

figure = plt.figure(figsize=(6, 6))
for c, m, data in [((16/255, 52/255, 166/255, 0.8), '^', pos_data), ((250/255, 218/255, 94/255, 1), 'o', neg_data)]:
    plt.scatter(data[:, 0], data[:, 1], c=c, marker=m)
plt.legend(['Positive class', 'Negative class'], loc='upper right')
plt.show()
````

````python
DATA_FILE = './data/data_10000_3dim.csv'
data_3d = pd.read_csv(DATA_FILE, sep=';').drop(['Unnamed: 0'], axis=1)

show_classes_ratio(data_3d)
````

````python
data_3d = np.array(data_3d)
pos_data = data_3d[data_3d[:, 3] == 1][:100]
neg_data = data_3d[data_3d[:, 3] == 0][:100]

figure = plt.figure(figsize=(8, 8))
ax = figure.gca(projection='3d')
for c, m, data in [((16/255, 52/255, 166/255, 0.8), '^', pos_data), ((250/255, 218/255, 94/255, 1), 'o', neg_data)]:
    ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=c, marker=m)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend(['Positive class', 'Negative class'], loc='upper right')
plt.show()
````
